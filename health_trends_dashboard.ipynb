# Nigeria Public Health Trends Dashboard
This project analyzes health indicators across Nigerian states from 2010 to 2023 using Python. The goal is to identify trends, disparities, and policy implications through data visualization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

%matplotlib inline
```

```python
df = pd.read_csv('nigeria_public_health.csv')
df.head()
```

```python
df.info()
df.describe()
```

## Life Expectancy Trends by State

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Life Expectancy (Years)', hue='State')
plt.title('Life Expectancy Over Time')
plt.ylabel('Years')
plt.show()
```

## Malaria Prevalence Trends

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Malaria Prevalence (%)', hue='State')
plt.title('Malaria Prevalence by State')
plt.ylabel('% Infected')
plt.show()
```

## Immunization Coverage by State (2023)

```python
latest_year = df[df['Year'] == 2023]
sns.barplot(data=latest_year, x='Immunization Coverage (%)', y='State', orient='h')
plt.title('Immunization Rates by State (2023)')
plt.xlabel('% Coverage')
plt.show()
```

## Correlation Between Health Metrics

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.drop(columns=['Year']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Interactive Life Expectancy Trend (Plotly)

```python
fig = px.line(df, x='Year', y='Life Expectancy (Years)', color='State',
              title='Interactive Life Expectancy by State')
fig.show()
```

## Summary of Insights
- Life expectancy is improving across all states.
- Malaria prevalence has decreased steadily in many regions.
- Immunization coverage is above 80% in most southern states.
- There is a positive correlation between health spending and life expectancy.
